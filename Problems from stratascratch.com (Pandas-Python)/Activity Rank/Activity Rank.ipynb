```python
import pandas as pd
```

```python
df = pd.read_csv('google_gmail_emails.txt',sep='\t')
```

```python
df=df.groupby('from_user')['to_user'].count().reset_index().rename(columns={'to_user':'n_of_emails_sent'}).sort_values('n_of_emails_sent',ascending=False)
df['rank']=df['n_of_emails_sent'].rank(method='first',ascending=False).astype(int)
```

```python
df=df.reset_index(drop=True)
df
```
