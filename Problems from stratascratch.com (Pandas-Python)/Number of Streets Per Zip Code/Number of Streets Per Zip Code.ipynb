```python
import pandas as pd
```

```python
df =pd.read_csv("sf_restaurant_health_violations.txt",sep='\t')
```

```python
df['street']=df['business_address'].apply(lambda x:x.split(' ')[1].lower() if str(x.split(' ')[0][0]).isdigit() else x.split(' ')[0].lower())
```

```python
result = df.groupby('business_postal_code')['street'].nunique().reset_index().sort_values(by='business_postal_code').rename(columns={'street':'n_street'}).sort_values(by='n_street',ascending=False)
```
